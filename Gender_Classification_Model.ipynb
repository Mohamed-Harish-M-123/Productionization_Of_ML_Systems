### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

### Hypothesis 1: Gender Distribution Difference Across Companies

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0): There is no significant difference in the gender distribution across different companies.**

**Alternative Hypothesis (H1): There is a significant difference in the gender distribution across different companies.**

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Create a contingency table
contingency_table = pd.crosstab(users_df['company'], users_df['gender'])

# Perform Chi-Square Test of Independence
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi2: {chi2}, p-value: {p}")

if p < 0.05:
    print("Reject the null hypothesis. There is a significant difference in gender distribution across companies.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in gender distribution across companies.")
````

##### Which statistical test have you done to obtain P-Value?

*I used the chi-square test for hypothesis testing to determine if the distribution of gender varies across different companies*

##### Why did you choose the specific statistical test?

*I chose the chi-square test because it is particularly effective for testing the independence of categorical variables. In this case, it helps determine if there is a significant difference in gender distribution across different companies, making it an appropriate choice for assessing whether the observed variations are statistically significant or occur by chance.*

### Hypothetical Statement - 2 Average Age Difference by Gender

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0): The average age of users does not differ significantly between genders.**

**Alternative Hypothesis (H1): The average age of users differs significantly between genders.**

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
male_ages = users_df[users_df['gender'] == 'male']['age']
female_ages = users_df[users_df['gender'] == 'female']['age']

# Perform t-Test
t_stat, p_val = stats.ttest_ind(male_ages, female_ages)

print(f"t-statistic: {t_stat}, p-value: {p_val}")

if p_val < 0.05:
    print("Reject the null hypothesis. The average age of users differs significantly between genders.")
else:
    print("Fail to reject the null hypothesis. The average age of users does not differ significantly between genders.")
````

##### Which statistical test have you done to obtain P-Value?

*I used a t-test to determine if there is a significant difference in the average age across different genders*

##### Why did you choose the specific statistical test?

*I chose the t-test because it is well-suited for comparing the means of two groups. In this case, it allows us to determine if the average age differs significantly between genders, making it the appropriate test for this hypothesis.*

### Hypothetical Statement - 3  Age Distribution Variation Across Companies

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**Null Hypothesis (H0): There is no significant variation in the age distribution across different companies.**

**Alternative Hypothesis (H1): There is a significant variation in the age distribution across different companies.**

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
groups = users_df.groupby('company')['age'].apply(list)

# Perform ANOVA
f_stat, p_val = stats.f_oneway(*groups)

print(f"F-statistic: {f_stat}, p-value: {p_val}")

if p_val < 0.05:
    print("Reject the null hypothesis. There is a significant variation in the age distribution across companies.")
else:
    print("Fail to reject the null hypothesis. There is no significant variation in the age distribution across companies.")
````

##### Which statistical test have you done to obtain P-Value?

*I used an F-test to perform a hypothesis test to check if the age distribution differs across different companies.*

##### Why did you choose the specific statistical test?

*I chose the F-test because it is suitable for comparing the variances of multiple groups. In this case, it helps determine if the age distribution significantly differs across different companies, making it the appropriate choice for this analysis.*

````python
def model_metrics(y_train, y_test, train_preds, test_preds):
    # Calculate metrics
    train_accuracy = accuracy_score(y_train, train_preds)
    test_accuracy = accuracy_score(y_test, test_preds)
    train_precision = precision_score(y_train, train_preds, average='weighted')  # Specify 'weighted' for multiclass
    test_precision = precision_score(y_test, test_preds, average='weighted')
    train_recall = recall_score(y_train, train_preds, average='weighted')
    test_recall = recall_score(y_test, test_preds, average='weighted')

    print(f"{'Train Accuracy':<20}{train_accuracy:.4f}")
    print(f"{'Test Accuracy':<20}{test_accuracy:.4f}")
    print(f"{'Train Precision':<20}{train_precision:.4f}")
    print(f"{'Test Precision':<20}{test_precision:.4f}")
    print(f"{'Train Recall':<20}{train_recall:.4f}")
    print(f"{'Test Recall':<20}{test_recall:.4f}")
    print("-"*50)

    # Confusion matrices
    train_confusion_matrix = confusion_matrix(y_train, train_preds)
    test_confusion_matrix = confusion_matrix(y_test, test_preds)

    # Plot confusion matrices
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.heatmap(train_confusion_matrix, annot=True, cmap='Blues', ax=axes[0], fmt="d")
    axes[0].set_xlabel('Predicted labels')
    axes[0].set_ylabel('True labels')
    axes[0].set_title('Train Confusion Matrix')
    sns.heatmap(test_confusion_matrix, annot=True, cmap='Blues', ax=axes[1], fmt="d")
    axes[1].set_xlabel('Predicted labels')
    axes[1].set_ylabel('True labels')
    axes[1].set_title('Test Confusion Matrix')

    plt.tight_layout()
    plt.show()
````

**For a positive business impact, the following evaluation metrics are often considered:**

**Accuracy:** *Provides a general measure of how well the model performs overall, indicating the proportion of correctly predicted instances out of the total instances. It is useful when the class distribution is balanced.*

**Precision:** *Measures the proportion of true positive predictions among all positive predictions made by the model. This is important when the cost of false positives is high, such as in fraud detection or medical diagnoses.*

**Recall (Sensitivity):** *Measures the proportion of true positive predictions among all actual positive instances. This metric is crucial when the cost of false negatives is high, such as in disease screening where missing a case could have serious consequences.*

**F1 Score:** *The harmonic mean of precision and recall. It provides a balanced measure that accounts for both false positives and false negatives, making it suitable when you need to balance precision and recall.*

**ROC-AUC:** *The area under the Receiver Operating Characteristic curve. This metric evaluates the model's ability to discriminate between positive and negative classes across different thresholds, providing insight into its performance across various scenarios.*

**Confusion Matrix:** *Provides a detailed breakdown of the true positives, true negatives, false positives, and false negatives. It helps in understanding the types of errors the model is making and can be useful for specific business requirements.*

**Why These Metrics?**

A*ccuracy is a good overall performance measure but may be misleading if classes are imbalanced.*

*Precision and Recall are crucial for evaluating models where the costs of false positives and false negatives are significant.*

*F1 Score balances precision and recall, offering a more comprehensive performance evaluation when dealing with imbalanced datasets.*

*ROC-AUC helps assess the model’s performance in distinguishing between classes under various threshold settings.*

*Confusion Matrix provides a detailed view of the model's performance, helping to identify specific areas for improvement.*
